```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.random.normal(0, 1, 100) * 10
```

```python
y = 2 * X + np.random.normal(0, 3, 100) +10
plt.plot(X, y , "o")
```

```python
# zagwar beldeh
# y= w*x + b
class LinearRegress():
  def __init__(self, learning_rate = 0.0001):
    self.w = np.random.randn(1)
    self.b = np.random.randn()
    self.learning_rate = learning_rate

  def forward(self, x):
    y_pred = self.w * x + self.b

    return y_pred

  def get_params(self):
    print("w: ", self.w)
    print("b: ", self.b)
    print("lr", self.learning_rate)

  # Optimization algorithm -- gradient descent
  # w_new <--  w- learning_rate * graw_w
  # b_new <--  b- learning_rate * graw_b
  def backward(self, grad_w, grad_b):
    self.w = self.w - self.learning_rate * grad_w
    self.b = self.b - self.learning_rate * grad_b

  def get_loss(self, x, y):
    y_pred = self.forward(x)
    # S = sum(y_pred - y)^2 / len(y)
    S = np.sum( (y_pred-y) ** 2) / len(x)
    return S

  def get_grad(self, x, y):
    y_pred = self.forward(x)
    # grad_w = (y_pred - y) * x
    grad_w = np.sum (x.dot(y_pred - y)) /len(x)
    # grad_b = y_pred - y
    grad_b = np.sum(y_pred - y) /len(x)
    return grad_w, grad_b
```

```python
model = LinearRegress(0.01)
plt.scatter(X,y)
x_1 = np.array([-25, 25])
y_1 = model.w * x_1 + model.b
plt.plot(x_1, y_1, 'r')
plt.show()
```

```python
# surgaya
print("gradient descentiin umnu")
print(X[:10])
print(model.forward(X)[:10])
model.get_params()
grad_w, grad_b = model.get_grad(X, y)
model.backward(grad_w, grad_b)
print("gradient descentiin umnu")
print(X[:10])
print(model.forward(X)[:10])
model.get_params()
```

```python
epoch = 100
model.get_params()
losses = list()
for i in range(epoch):
  grad_w, grad_b = model.get_grad(X, y)
  model.backward(grad_w, grad_b)
  loss = model.get_loss(X, y)
  losses.append(loss)

  if (i % 10 == 0):
    print(i, loss)
    plt.scatter(X,y)
    x_1 = np.array([-25, 25])
    y_1 = model.w * x_1 + model.b
    plt.plot(x_1, y_1, 'r')
    plt.show()
model.get_params()
```

```python
plt.plot(range(epoch), losses)
plt.ylabel("Loss")
plt.xlabel("Epoch #")
plt.show()
```
